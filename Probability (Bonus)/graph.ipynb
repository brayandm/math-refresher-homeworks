```python
import matplotlib.pyplot as plt
import numpy as np

def plot_grid(n, interval):
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # Set light gray background
    ax.set_facecolor('#D3D3D3')
    
    # Draw horizontal and vertical lines to form the grid
    for i in range(0, n+1, interval):
        ax.axhline(i, color='gray', linewidth=0.5)
        ax.axvline(i, color='gray', linewidth=0.5)
    
    # Set plot limits and aspect ratio
    ax.set_xlim(0, n)
    ax.set_ylim(0, n)
    ax.set_aspect('equal')
    
    # Set axis ticks in intervals of 5
    ax.set_xticks(range(0, n+1, interval))
    ax.set_yticks(range(0, n+1, interval))

    # Define lines
    x = np.linspace(0, n, 1000)
    y1 = x
    y2 = x - 3
    y3 = x + 15
    
    # Plot lines
    ax.plot(x, y1, '-k', label='y=x')
    ax.plot(x, y2, '-y', label='y=x-3')
    ax.plot(x, y3, '-b', label='y=x+15')

    # Fill between lines
    ax.fill_between(x, y1, y3, where=(y1 <= y3), color='#6fa8dc', interpolate=True)
    ax.fill_between(x, y1, y2, where=(y1 >= y2), color='#fffdaf', interpolate=True)
    
    plt.legend()
    plt.show()

# Plot a 60x60 grid with the described lines and shaded regions
plot_grid(60, 5)
```
